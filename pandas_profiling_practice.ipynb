# Pandas Profiling

- Pandas profiling is a open source Exploratory Data Analysis tool which is perfom data analysis of the given data frame within few lines of code
- It can perform analysis like co-relation, missing values, mean etc.
- It generates a well designed html file which consists of all these analysis

### Instaling Pandas Profiling

```python
pip install pandas-profiling
```

### Importing Necessary Libraries

```python
import pandas as pd
import pandas_profiling as pp
```

### If you get any error like 'NoSuchPackage' or something like that try running this code

```python
# import sys
# !{sys.executable} -m pip install pandas-profiling
```

### Importing Data set

```python
df = pd.read_csv('student_data.csv')
df.head()
```

### Using Pandas Profiling

```python
# creating a variable to store pandas profile
profile = pp.ProfileReport(df)
profile.to_file('output.html')
```

### Output file
- We'll get an html file named 'output' (Because i named it as 'output', you can name it anything) in the folder where your current ipynb is running.
- Now we're just a click away to explore the awesomeness of pandas profiling.
